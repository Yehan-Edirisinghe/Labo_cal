```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def cauchy(x,M,L):
    return L / ( np.pi * ((x-M)**2 + L**2) )
```

```python
M,L = 4,3

x = np.linspace(0,10,100)
y = [cauchy(i,M,L) for i in x]

plt.plot(x,y)
plt.show()
```

```python
def f_cauchy(M,L,x_size):

    y_max = cauchy(M,M,L)

    #generate points randomly on the plane
    x , y = np.random.uniform(M-x_size, M+x_size) , np.random.uniform(0, y_max)
    
    #search until a point is inside the function
    while y > cauchy(x,M,L):

        #change point if not adeguate
        x , y = np.random.uniform(M-x_size, M+x_size) , np.random.uniform(0, y_max)
    
    return x
```

```python
M,L = 2,1
data_1 = [f_cauchy(M,L,3*1*L) for j in range(1000)]
data_2 = [f_cauchy(M,L,3*20*L) for j in range(1000)]
data_3 = [f_cauchy(M,L,3*50*L) for j in range(1000)]
data_4 = [f_cauchy(M,L,3*100*L) for j in range(1000)]
```

```python
plt.hist(data_4,bins=15)
plt.hist(data_3,bins=15)
plt.hist(data_2,bins=15)
plt.hist(data_1,bins=15)

plt.show()
```

```python
mean = np.average(data_1)
sigma = np.std(data_1)

print(mean,sigma)
```
