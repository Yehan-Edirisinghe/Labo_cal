```python
from Libraries.stats import*
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gen(N,xmin,xmax):
    return [rand_uniform(xmin,xmax,N) for i in range(N)]
```

```python
N,xmin,xmax = 4000,1,4
data = gen(N,xmin,xmax)
# data = toy_Gauss(0,1).generate()
plt.hist(data)
plt.show()
```

```python
def count(xmin,xmax,data):
    counter = 0
    for i in data:
        if i>xmin and i<=xmax:
            counter += 1
    return counter
```

```python
def binning(k,data):
    xmin = np.min(data)
    xmax = np.max(data)
    delta = (xmax-xmin)/k
    binned_content = np.zeros(k)

    for i in range(k):
        binned_content[i] = count((xmin+i*delta),(xmin+(i+1)*delta),data)

    return binned_content,range(k)
```

```python
nbins = 100
bin_content,bin_edges = binning(nbins,data)
plt.scatter(bin_edges,bin_content)
plt.show()
```

```python
m,v,s,k = stat(bin_content)
print("mean: %f     \nVariance: %f" % (m,v))
```
