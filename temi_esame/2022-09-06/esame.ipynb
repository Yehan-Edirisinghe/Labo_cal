```python
import numpy as np
import matplotlib.pyplot as plt
from Libraries.funcs import *
from Libraries.stats import*
```

```python
def parabolic(x,a,c):
    return a*np.power(x,2)+c
```

```python
def f(x,x0):
    
    a = -3/(4*np.power(x0,3))
    c = 3/(4*x0)
    if x < -x0 or x > x0:
        return 0
    return parabolic(x,a,c)
```

```python
x0 = 3
x = np.linspace(-5,5,100)
y = [f(i,x0) for i in x]

plt.plot(x,y)
plt.show()
```

```python
class lib:
    def __init__(self,x0) -> None:
        self.x0 = x0
        self.a = -3/(4*np.power(self.x0,3))
        self.c = 3/(4*self.x0)
    def eval(self,x):
        return parabolic(x,self.a,self.c)
```

```python
def TAC_f(N,x0,ymax):
    data = []
    p = lib(x0)
    a,c = p.a, p.c

    for i in range(N):
        x = np.random.uniform(-x0,x0)
        y = np.random.uniform(0,ymax)
        if parabolic(x,a,c) > y:
            data.append(x)
    return data
```

```python
x0 = 3
xmax,ymax = max_sez_aurea(f,-x0,x0,x0)
data = TAC_f(1000,x0,ymax)
plt.hist(data,bins=sturges(data),density=True)
plt.show()
```

```python
m,v,s,k = stat(data)
print("Mean: %f  Sigma: %f" % (m,np.sqrt(v)))
```
